```python
import pandas as pd
```

```python
#csv path to raw data
csv_path1 = "raw_data/schools_complete.csv"
csv_path2 = "raw_data/students_complete.csv"
schools_df = pd.read_csv(csv_path1, encoding='iso-8859-1', low_memory=False)
students_df = pd.read_csv(csv_path2, encoding='iso-8859-1', low_memory=False)
```

```python
schools_table= schools_df.rename(columns={"name":"School Name"})
students_table= students_df.rename(columns={"school":"School Name"})
```

```python
#merge tables 
merged_table= pd.merge(schools_table, students_table,on="School Name",how="outer")
merged_table.head(1)
math_count = 0
reading_count = 0
math_list = []
reading_list = []
total_schools=schools_table["School ID"].count()
total_student= students_table["name"].count()
total_budget=schools_table["budget"].sum()
average_math=students_table["math_score"].mean()
average_reading=students_table["reading_score"].mean()

#for loop for math % passing and counting number of passing 
for x in merged_table["math_score"]:
    if x>=70:
        math_count =  math_count+1
        counter =  1
    else:
        counter = 0
    math_list.append(counter)
    counter = 0
merged_table["math"] = math_list
passing_math=(math_count/total_student*100)

#for loop for math % passing and counting number of passing
for x in students_table["reading_score"]:
    if x>=70:
        reading_count = reading_count + 1
        counter = 1
    else:
        counter = 0
    reading_list.append(counter)
    counter = 0
merged_table["read"] = reading_list
passing_reading=(reading_count/total_student*100)

print(passing_reading)

passing_math=(math_count/total_student*100)

passing_math
```

```python
overall_passing = (passing_math+passing_reading)/2
overall_passing
```

```python
district_summary = pd.DataFrame({"total schools": total_schools,
                                "total students": total_student,
                                 "total budget": total_budget,
                                 "average math scores": average_math,
                                 "average reading scores": average_reading,
                                 "passing math scores": passing_math,
                                 "passing reading scores": passing_reading,
                                 "overall passing rate": overall_passing
                                },index=[0])
district_summary
```

```python
school_summary = merged_table.groupby(["School Name", "type"])
print(school_summary)
school_summary.count()


students = school_summary["name"].count()
school_budget = school_summary["budget"].mean()
per_student_bud = school_budget/students
average_math = school_summary.mean

```

```python
group_table =  merged_table.groupby(["School Name", "type"])


totalbudget = group_table["budget"].mean()
average_math = group_table["math_score"].mean()
average_reading = group_table["reading_score"].mean()
total_passing_math = group_table["math"].sum()
total_passing_read = group_table["read"].sum()
percent_pass_math = (total_passing_math/students *100)
percent_pass_read = (total_passing_read/students *100)
overall_passing = (percent_pass_math + percent_pass_read)/2



school_summary = pd.DataFrame({
                               "school type": type,
                               "total students": students,
                               "total school budget": totalbudget,
                               "per student budget": per_student_bud,
                               "average math score": average_math,
                                "average reading score": average_reading,
                               "% passing math": percent_pass_math,
                               "% passing reading": percent_pass_read,
                               "overall % passing rate": overall_passing,
                                })
school_summary.head(10)
```

```python
top_schools = school_summary.sort_values("overall % passing rate", ascending=False)
top_schools.head(1)
```

```python
bottom_schools = school_summary.sort_values("overall % passing rate", ascending=True)
bottom_schools.head(1)
```
